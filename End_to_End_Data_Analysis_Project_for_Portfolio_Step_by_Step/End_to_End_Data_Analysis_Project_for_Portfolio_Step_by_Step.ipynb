**Hypothesis 1:**

```python
cat_profit = data.groupby('Category')['Profit'].sum()
```

```python
cat_profit.plot(kind='bar')
plt.title("Profit By Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
#This hypothesis is supported as technology products have the highest profit margin of the three Category
```

**Hypothesis 2:**

```python
reg_sales = data.groupby('Region')['Sales'].sum()
```

```python
reg_sales
```

```python
reg_sales.plot(kind='bar')
plt.title("Sales By Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
#Conclusion: This hypothesis is not supported as Central Region has the highest sales
```

**Hpothesis 3:**

```python
data.columns
```

```python
data['order month']=pd.DatetimeIndex(data['Order Date']).month
```

```python
month_sales=data.groupby('order month')['Sales'].sum()
```

```python
month_sales
```

```python
month_sales.plot(kind='line')
plt.title("Total Sales By Month")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
#Conclusion: This hypothesis is supported as sales are higher during certain months of the year
```

**Hypothesis 4:**

```python
total_oders_by_shippng_mode = data.groupby('Ship Mode').size()
```

```python
total_oders_by_shippng_mode
```

```python
returned_order_by_shipping_mode = data[data['Profit']<0].groupby('Ship Mode').size()
```

```python
returned_per_by_shipping_mode = (returned_order_by_shipping_mode/total_oders_by_shippng_mode)*100
```

```python
returned_per_by_shipping_mode
```

```python
returned_per_by_shipping_mode.plot(kind='bar')
plt.title("Return Per By Shipping Mode")
plt.xlabel("Shipping Mode")
plt.ylabel("Return Per")
plt.show()
#Conclusion: This hypothesis is supported as orders with same day shipping have the lowest rate of returned product
```

**Hypothesis 5:**

```python
data['Order Day']=pd.DatetimeIndex(data['Ship Date']).day_name()
```

```python
day_profit = data.groupby('Order Day')['Profit'].sum()
day_profit
```

```python
day_profit.plot(kind='bar')
plt.title("Total Profit By Day of the week")
plt.xlabel("Day of the week")
plt.ylabel("Total Profit")
plt.show()
#Conclusion: This hypothesis is supported as Friday has the highest profit margin
```
