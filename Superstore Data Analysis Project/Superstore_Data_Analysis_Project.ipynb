**Exploratory Data Analysis**

**Count of Sub-Category**

```python
data.columns
```

```python
col = 'Sub-Category'
sns.countplot(x='Sub-Category', data = data)
plt.figure(figsize=(10,5))
plt.xticks(rotation=60)
plt.figure(figsize=(10,5))
plt.title('Count of SubCategory')
sns.barplot(data=data,x=data.index,y=col,palette= 'Blues')
plt.show()
```

**Best Performing Category**

```python
sns.countplot(x='Category', data=data)
plt.title('Count of Category')
plt.show()
```

```python
df2 = pd.DataFrame(data.groupby(['Segment'])[['Profit']].mean())
df2
```

```python
sns.set_theme(style="whitegrid")
sns.barplot(data = df2, x = df2.index, y = df2.Profit, palette = "Blues")
plt.title("Customer Segment Profitability")
plt.show()
```

```python
data.columns
```

```python
df_stackb = data[['Ship Mode','Discount','Profit']]
df_stackb = df_stackb.groupby(['Ship Mode']).sum().reset_index()
plt.figure(figsize=[10,15])
plt.subplot(3,1,1)
plt.bar(x=df_stackb['Ship Mode'], height=df_stackb['Discount'], color='skyblue')
plt.bar(x=df_stackb['Ship Mode'], height=df_stackb['Profit'], bottom=df_stackb['Discount'], color='green')
plt.title("Sales & Profit Across Ship Modes", size=20, pad=20)
plt.legend(['Sales','Profit'])
```

```python
region_analysis = pd.DataFrame(data.groupby(['Region'])['Profit'].sum().reset_index())
region_analysis
```

```python
plt.pie(region_analysis.Profit,
        labels = region_analysis.Region,
        startangle = 90, autopct = "%1.0f%%",
        explode = [0, 0, 0, 0.1])

plt.title("Most Profitable by Region")
plt.show()
```
