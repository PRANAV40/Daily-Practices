**Data Cleaning  (Replace '--'  to NaN)**

```python
import numpy as np
```

```python
df = df.replace('--',np.nan,regex=True)
```

```python
df.head(20)
```

**Check Null Values In The Dataset**

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
per_missing = df.isnull().sum() * 100 / len(df)
```

```python
per_missing
```

```python
sns.heatmap(df.isnull())
```

```python
df.dropna(axis=0,inplace=True)
```

```python
sns.heatmap(df.isnull())
```

**Data Cleaning [ Rank Column ]**


```python
df['Rank']
```

```python
df.dtypes
```

```python
df['Rank'] = df['Rank'].str[0:-2]
```

```python
df.tail()
```

```python
df['Rank'] = df['Rank'].str.replace(',','').astype('int')
```

```python
df.dtypes
```

**Data Cleaning [ Video Uploads & Subscribers ]**

```python
df.dtypes
```

```python
df['Video Uploads'] = df['Video Uploads'].astype('int')
```

```python
df.dtypes
```

```python
df['Subscribers'] = df['Subscribers'].str.replace(',','').astype('int')
```

```python
df.dtypes
```

**Data Cleaning [ Grade Column ]**

```python
df['Grade'].unique()
```

```python
df['Grade'] = df['Grade'].map({'A++ ':5,'A+ ':4,'A ':3,'A- ':2,'B+ ':1})
```

```python
df.dtypes
```

**Find Average Views For Each Channel**

```python
df.columns
```

```python
df['Avg_Views'] = df['Video views']/df['Video Uploads']
```
