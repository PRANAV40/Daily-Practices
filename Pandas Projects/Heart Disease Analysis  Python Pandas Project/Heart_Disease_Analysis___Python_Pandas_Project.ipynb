**20. Plot Continuous Variables**

```python
create_val = []
count_val = []

for column in data.columns:
  if data[column].nunique() <=10:
    create_val.append(column)
  else:
    count_val.append(column)
```

```python
create_val
```

```python
count_val
```

```python
data.hist(count_val,figsize=(15,6))
plt.tight_layout()
plt.show()
```
