**In Which Month Maximum Number of Forest Fires Were Reported?**

**Display Number of Fires Were Reported In Amazonas (Day-Wise)**

```python
df6 = df[df['state'] == 'Amazonas']
```

```python
df6.loc[:, 'date'] = pd.to_datetime(df6['date'])
```

```python
df6 = df6.copy()
df6['date'] = pd.to_datetime(df6['date'])
```

```python
day = df6.groupby(df6['date'].dt.dayofweek).sum(numeric_only=True).number
```

```python
import calendar
day.index = [calendar.day_name[i] for i in range(0,7)]
day = day.reset_index()
```

```python
day
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='index',y='number',data=day)
```

**Find Total Number of Fires  Were Reported In 2015 And Visualize Data Based on Each ‘Month’**

```python
fire = df[df['year']==2015].groupby('month')['number'].sum().reset_index()
```

```python
fire
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='month',y='number',data=fire)
```

**Find Average Number of Fires Were Reported From Highest to Lowest (State-Wise)**

```python
df8 = df.groupby('state')['number'].mean().sort_values(ascending=False).reset_index()
```

```python
df8
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='state',y='number',data=df8)
plt.xticks(rotation=50)
plt.show()
```
