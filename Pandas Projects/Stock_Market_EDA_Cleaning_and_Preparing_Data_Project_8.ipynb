```python
def cap_classify(value):
  if value<=31020.77:
    return "Smallcap"
  elif 31020.77<value<=94438.88:
    return "Mediumcap"
  else:
    return "Largecap"

stock_df['Cap']=stock_df['market_cap'].apply(cap_classify)
```
