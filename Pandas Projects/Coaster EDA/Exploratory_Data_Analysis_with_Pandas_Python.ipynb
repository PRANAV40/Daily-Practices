# **Data Preperation**
*   Dropping irrelevant columns and rows
*   Identifying duplicated columns

```python
data = data[['coaster_name','Location', 'Status','Manufacturer','year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean','speed_mph', 'height_ft','Inversions_clean', 'Gforce_clean']].copy()
```

```python
data['opening_date_clean'] = pd.to_datetime(data['opening_date_clean'])
```

**Rename the columns**

```python
data = data.rename(columns={'coaster_name':'Coaster_Name',
                     'year_introduced':'Year_Introduced',
                     'opening_date_clean':'Opening_Date',
                     'speed_mph':'Speed_mph',
                     'height_ft':'Height_ft',
                     'Inversions_clean':'Inversion',
                     'Gforce_clean':'Gforce'})
```

What are the locations with the fastest roller coasters (minimum of 10)?

```python
data.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean', 'count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean']\
    .plot(kind='barh',figsize=(12,8))
plt.title('Average Coast Speed by Loaction')
plt.xlabel('Average Speed (mph)')
plt.ylabel('Location')
plt.show()
```
